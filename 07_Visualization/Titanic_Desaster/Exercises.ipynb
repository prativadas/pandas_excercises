### Step 1. Import the necessary libraries

```python
import pandas as pd

import numpy as np

import seaborn as sns

import matplotlib.pyplot as plt


%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv)

### Step 3. Assign it to a variable titanic 

```python
titanic = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv', sep=',')


titanic.head()
```

### Step 4. Set PassengerId as the index 

```python
titanic = titanic.set_index('PassengerId')
```

```python
titanic.head()
```

### Step 5. Create a pie chart presenting the male/female proportion

https://stackabuse.com/matplotlib-pie-chart-tutorial-and-examples

```python
titanic.Sex.value_counts().plot.pie(colors= ['tab:orange', 'tab:pink'], autopct='%.1f%%')

plt.show()
```

```python
x= titanic.Sex.value_counts()

labels = ['male', 'female']

colors = ['tab:red', 'tab:cyan']

explode = [0,0.1]      # used for highlighing a portion


fig, ax = plt.subplots()

ax.pie(x, labels = labels, colors = colors, autopct='%.0f%%', explode= explode)

ax.set_title('male/female proportion')

plt.show()
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
titanic.plot.scatter(x= 'Age' , y = 'Fare', c='cyan')
```

```python
sns.scatterplot(data=titanic, x= 'Age' , y = 'Fare')
```

```python
g = sns.FacetGrid(titanic, hue='Sex')

g.map_dataframe(sns.scatterplot, x="Age", y="Fare",)

g.add_legend()

g.set_axis_labels("Total bill", "Tip")
```

```python
g = sns.FacetGrid(titanic, hue='Sex')

g.map_dataframe(sns.scatterplot, x="Age", y="Fare",)

g.add_legend()

g.set_axis_labels("Total bill", "Tip")

```

```python

plt.figure(figsize=(5,5))

sns.scatterplot(x=titanic.Age, y=titanic.Fare, hue=titanic.Sex, alpha=.8,) 


plt.show() 
```
